```
#Binomial Distribution
#Binomial distribution
#To find the probability of having four or less correct answers by random attempts, we apply the function dbinom with 
#x = 0,1,2,3,4. or the pbinom function with x = 4
db = pbinom(4, size=12, prob=0.2)
cat('The probability is: ', db)
```

```
#Poisson's Distribution
#The probability of having seventeen or more cars crossing the bridge in a minute is in the u pper tail of the probability density function.
prob= ppois(16, lambda=12, lower=FALSE)	# upper tail
cat('The Probability is : ', prob)
```

```
#Continuous uniform distribution

#We apply the generation Functiun r'unif ot the uniform distribution to generate ten random numbers between one and three.
ran = runif(10, min=1, max=3)
cat('The random numbers are 2', ran)
```

```
#exponential Distribution
#We can apply the Function pexp at the exponential distribution with rate=i/3 to find out the probability

p = pexp(2, rate=1/3)
cat('The probability is :',p)
```

```
#normal Distribution
#since we are looking for the percentage of students scoring higher than 54, we are interested in the upper tail of the normal distributiunt

norm = pnorm(84, mean= 72, sd=15.2, lower.tail=FALSE)
cat('The probability is', norm)
```

```
#Chi-squared distribution
#We apply the quantile function qchisq of the Chi-Squared distribution against the decimal values 8.95.

per = qchisq(0.95, df=7) # 7 degrees of freedom
cat('The 95th percentile is ', per)
```

```
#student t distribution
#We apply the quantile function qt of the Student t distribution against the decimal values 8.615 and 9.975.

quant = qt(c(.025, 0.975), df=5) # 5 degrees of treedom
cat('The 2.5th and 97.5th percentiles of the distribution are : ', quant[1],",", quant[2] )
```

```
#F Distribution
#We apply the quantile function qt of the F distribution against the decimal value 9.95.
q = qf(.95, df1=5, df2=2)
cat('The 95th percentile of the distribution is : ', q)
```

```
set.seed(3000) 
xseq<- seq(-4,4,.01) 
densities<- dnorm(xseq, 0,1) 
cumulative<- pnorm(xseq, 0, 1) 
randomdeviates<- rnorm(1000,0,1) 
par(mfrow=c(1,3), mar=c(3,4,4,2)) 
plot(xseq, densities, col="darkgreen",xlab="", ylab="Density", type="l",lwd=2, cex=2, main="PDF of Standard Normal", cex.axis=.8) 
```

```
plot(xseq, cumulative, col="darkorange", xlab="", ylab="Cumulative Probability",type="l",lwd=2, cex=2, main="CDF of Standard Normal", cex.axis=.8) 
```

```
hist(randomdeviates, main="Random draws from Std Normal", cex.axis=.8, xlim = c(4,4))
```
