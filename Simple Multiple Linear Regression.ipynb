SIMPLE LINEAR REGRESSION ANALYSIS



```
library(ggplot2)  #including ggplot library for mpg dataset and data analysis
```

```
df<-mpg   #making copy of the dataset 
```

```
head(df)  #displaying the first 6 values of mpg
```

```
city<-df$cty  #copying cty column to city 
```

```
disp<-df$displ  #copying displ column to disp 
```

```
qplot(disp,city)

#scatter plot showing relation between disp and city
```

```
model=lm(city~disp,data=df)

# linear model for equation 
```

```
coefficients(model)

#slope(m) and intercept(c) of the best fit line
```

```
ggplot(df,aes(disp,city))+geom_point()+geom_smooth(method="lm")

#getting best fit line by using the linear regression model
```

```
summary(model)

#summarizing model
```

```
a=data.frame(disp=c(4,6,7))
```

```
predict(model,a)
#predicting the model through different values to check efficiency 
```

```
yhat=fitted(model)

#model prediction on the mpg dataset
```

```
residual=city-yhat

#residual of the model
```

```
residual
```

```
residuals(model)
```

MULTIPLE LINEAR REGRESSION ANALYSIS

```
df1<-mpg   #copying the mpg dataset into df1
```

```
head(df1)  
```

```
cor(df1$cty,df1$hwy)  #correlation between cty and hwy
```

```
cor(df1$cty,df1$year) #correlation between cty and year
```

```
cor(df1$cty,df1$cyl) #correlation between cty and cyl
```

```
cor(df1$cty,df1$displ) #correlation between cty and displ
```

```
y<-df$cty
x1<-df1$displ
x2<-df1$cyl
x3<-df1$hwy

#storing the predictors and response column in variables for multiple regression analysis
```

```
qplot(x1,y)

#Scatter plot between disp and cty 






```

```
ggplot(df1,aes(x1,y))+geom_point()+geom_smooth(method="lm")

#ggplot lm plot between disp and cty 
```

```
qplot(x2,y)

#Scatter plot between cyl and cty 
```

```
ggplot(df,aes(x2,y))+geom_point()+geom_smooth(method="lm")

#ggplot lm plot between cy1 and cty 
```

```
qplot(x3,y)

#Scatter plot between hwy and cty 
```

```
ggplot(df1,aes(x3,y))+geom_point()+geom_smooth(method="lm")

#ggplot lm plot between hwy and cty
```

```
model1=lm(y~x1+x2+x3)

#generating multiregression model on the dataset 
```

```
model1
```

```
coefficients(model1)

#slope(m) and intercept(c) of the best fit line
```

```
summary(model1)

#summarizing the model
```

```
residuals(model1)

#residual value for the dataset
```
